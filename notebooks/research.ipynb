# SMA Crossover Strategy Research

This notebook is designed to run inside the LEAN Research Environment.

Use the `qb` variable to access the QuantBook API.
See: https://www.quantconnect.com/docs/v2/research-environment

```python
qb = QuantBook()
spy = qb.AddEquity("SPY", Resolution.Daily).Symbol

history = qb.History(spy, 200, Resolution.Daily)
history['fast_sma']  = history['close'].rolling(20).mean()
history['med_sma']   = history['close'].rolling(50).mean()
history['slow_sma']  = history['close'].rolling(100).mean()

history[['close','fast_sma','med_sma','slow_sma']].plot(figsize=(14,6), title='SPY SMA Crossover')
```

```python
# ATR calculation
atr = history['close'].diff().abs().rolling(14).mean()
atr.plot(figsize=(14,3), title='SPY Daily ATR(14)')
```
